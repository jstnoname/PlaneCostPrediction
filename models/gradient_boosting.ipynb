реализую данных алгоритмы, а после сравню

```python
import mlflow
import optuna
from mlflow.utils.autologging_utils import disable_autologging
from optuna import Trial
from pyspark.ml import Pipeline, PipelineModel
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.regression import GBTRegressor
from pyspark.sql import SparkSession, DataFrame

from utility_data.utils import time_train_test_split, wape_metric, Config, read_parquet
```

```python
config = Config.load_config()
spark = SparkSession \
    .builder \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.shuffle.partitions", "5") \
    .getOrCreate()

mlflow.set_tracking_uri(config.dagshub_uri)
mlflow.set_experiment('PlaneCostPrediction')
mlflow.pyspark.ml.autolog(log_models=False, log_datasets=False)
mlflow.lightgbm.autolog(log_models=False, log_datasets=False)

df = read_parquet(spark, config.hf_dataset, config.hf_read_token, batch_size=10000, partition_size=5)

df.show(5)
```

```python
def spark_objective(trial: Trial, model_name: str, model_cols: dict[str, str], train_data: DataFrame, valid_data: DataFrame):
    params = {
        'maxDepth': trial.suggest_int('maxDepth', 2, 8),
        'maxBins': trial.suggest_int('maxBins', 66, 100),
        'minInfoGain': trial.suggest_float('minInfoGain', 1e-6, 1e-1, log=True),
        'maxIter': trial.suggest_int('maxIter', 10, 75),
        'stepSize': trial.suggest_float('stepSize', 1e-6, 1e-1, log=True),
        'subsamplingRate': trial.suggest_float('subsamplingRate', 0.6, 1.0),
        'checkpointInterval': 20
    }

    with disable_autologging():
        optuna_model = GBTRegressor(featuresCol=model_cols['featuresCol'], labelCol=model_cols['labelCol'], predictionCol=model_cols['predictionCol'], **params, seed=config.random_state)
        optuna_model = optuna_model.fit(train_data)
    val_preds = optuna_model.transform(valid_data)

    val_wape = wape_metric(val_preds, label_col=model_cols['labelCol'], prediction_col=model_cols['predictionCol'])
    val_rmse = RegressionEvaluator(predictionCol=model_cols['predictionCol'], labelCol=model_cols['labelCol']).evaluate(val_preds)

    mlflow.log_metric(f'{model_name}_val_rmse', val_rmse, step=trial.number)
    mlflow.log_metric(f'{model_name}_val_wape', val_wape, step=trial.number)

    return val_wape
```

```python
def spark_train(dataframe: DataFrame):
    models = {
        'economy': {'labelCol': 'target_economy', 'featuresCol': 'features', 'predictionCol': 'prediction_economy'},
        'comfort': {'labelCol': 'target_comfort', 'featuresCol': 'features', 'predictionCol': 'prediction_comfort'},
        'business': {'labelCol': 'target_business', 'featuresCol': 'features', 'predictionCol': 'prediction_business'},
    }

    train, test, train_meta = time_train_test_split(dataframe, test_size=0.25)
    valid, test, valid_meta = time_train_test_split(test, test_size=0.5)

    cat_cols = [name for name, dtype in dataframe.dtypes if dtype == 'string']
    index_cols = [f'{name}_index' for name in cat_cols]
    exclude_cols = ['target_economy', 'target_comfort', 'target_business'] + cat_cols + ['departure_month', 'departure_day']

    indexer = StringIndexer(inputCols=cat_cols, outputCols=index_cols, handleInvalid='keep')
    assembler = VectorAssembler(inputCols=[col for col in df.columns if col not in exclude_cols] + index_cols, outputCol='features')

    with mlflow.start_run(run_name='spark_gradient_boosting', tags={'mlflow.user': 'Alexander'}) as run:
        mlflow.log_params({
            "all_days": train_meta["all_days"],
            "train_days_count": train_meta["train_size"],
            "valid_days_count": valid_meta["train_size"],
            "test_days_count": valid_meta["all_days"] - valid_meta["train_size"]
        })

        feature_pipeline = Pipeline(stages=[indexer, assembler]).fit(train)

        train_piped = feature_pipeline.transform(train).cache()
        valid_piped = feature_pipeline.transform(valid).cache()

        model_stages = []
        for cur_model_name, cur_model_cols in models.items():
            study = optuna.create_study(study_name=f'Optuna_{cur_model_name}_study', direction='minimize')
            study.optimize(lambda trial: spark_objective(trial, cur_model_name, cur_model_cols, train_piped, valid_piped), n_trials=10, show_progress_bar=True)

            print(f"--> {cur_model_name} val WAPE = {study.best_value * 100:.2f}%")

            with disable_autologging():
                model = GBTRegressor(
                    featuresCol=cur_model_cols['featuresCol'], labelCol=cur_model_cols['labelCol'], predictionCol=cur_model_cols['predictionCol'], **study.best_params, seed=config.random_state
                ).fit(train_piped)
                model_stages.append(model)

        train_piped.unpersist()
        valid_piped.unpersist()

        model_pipeline = PipelineModel(stages=[*feature_pipeline.stages, *model_stages])
        mean_rmse, mean_wape = 0, 0
        preds = model_pipeline.transform(test).cache()

        for cur_model_name, cur_model_cols in models.items():
            rmse = RegressionEvaluator(predictionCol=cur_model_cols['predictionCol'], labelCol=cur_model_cols['labelCol']).evaluate(preds)
            wape = wape_metric(preds, label_col=cur_model_cols['labelCol'], prediction_col=cur_model_cols['predictionCol'])
            print(f'{cur_model_name} predictions\nRMSE = {rmse:.4f}\nWAPE = {wape * 100:.4f}%\n')

            mean_rmse += rmse
            mean_wape += wape

            mlflow.log_metric(f'{cur_model_name} RMSE', rmse)
            mlflow.log_metric(f'{cur_model_name} WAPE', wape)

        mean_rmse /= len(models)
        mean_wape /= len(models)

        print(f'Mean RMSE = {mean_rmse:.4f}\nMean WAPE = {mean_wape * 100:.4f}%\n')

        mlflow.log_metric('Mean RMSE', mean_rmse)
        mlflow.log_metric('Mean WAPE', mean_wape)
        mlflow.spark.log_model(model_pipeline, 'spark_gb')
```

Перед началом обучение градиентного бустинга из других библиотек нужно определиться какие библиотеки мы будем использовать.
Т.к. моя задача так же научиться эффективно управлять памятью, то библиотека `catboost` не подходит. Она хорошо работает с категориальными признаками и устойчива к переобучению, однако из-за постороения
симметричных деревьев catboost требует намного больше RAM и обучается дольше, чем другие библиотеки. `xgboost` строит деревья поуровнево, из-за чего может строить узлы там, где ошибка уже минимальна.
Поэтому для реализации градиентого бустинга буду использовать `lightgbm`

```python
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_percentage_error, root_mean_squared_error
from sklearn.pipeline import Pipeline as SklearnPipeline
from pandas import DataFrame

from utility_data.models import MultimodelLGMB, DataPreprocessor
```

```python
def lightgbm_objective(trial: Trial, model_name: str, X_train: DataFrame, y_train: DataFrame, X_val: DataFrame, y_val: DataFrame):
    params = {
        'objective': 'mse',
        'metric': 'rmse',
        'learning_rate': trial.suggest_float('learning_rate', 1e-6, 1e-1, log=True),
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'num_leaves': trial.suggest_int('num_leaves', 10, 1000),
        'max_depth': trial.suggest_int('max_depth', 1, 12),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'random_state': config.random_state,
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-3, 10.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-3, 10.0, log=True),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'verbose': -1
    }

    y_train_smooth = y_train + 1

    with disable_autologging():
        optuna_model = LGBMRegressor(**params)
        optuna_model.fit(X_train, y_train_smooth)
    val_preds = optuna_model.predict(X_val) - 1
    val_preds = val_preds.clip(0)

    val_wape = mean_absolute_percentage_error(y_val, val_preds, sample_weight=y_val)
    val_rmse = root_mean_squared_error(y_val, val_preds)

    mlflow.log_metric(f'{model_name}_val_rmse', val_rmse, step=trial.number)
    mlflow.log_metric(f'{model_name}_val_wape', val_wape, step=trial.number)

    return val_wape
```

```python
def lightgbm_train(dataframe: DataFrame):
    models = {
        'economy': 'target_economy',
        'comfort': 'target_comfort',
        'business': 'target_business',
    }

    train, test, train_meta = time_train_test_split(dataframe, test_size=0.25)
    valid, test, valid_meta = time_train_test_split(test, test_size=0.5)
    train, valid, test = train.toPandas(), valid.toPandas(), test.toPandas()

    cat_cols = [name for name, dtype in dataframe.dtypes if dtype == 'string']
    targets = ['target_economy', 'target_comfort', 'target_business']
    service_cols = ['departure_month', 'departure_day']     # исключаю эти колонки, чтобы не было переобучения по этому признаку
    exclude_cols = targets + service_cols

    X_train_optuna, X_valid_optuna = train.drop(columns=exclude_cols), valid.drop(columns=exclude_cols)
    for col in cat_cols:
        X_train_optuna[col] = X_train_optuna[col].astype('category')
        X_valid_optuna[col] = X_valid_optuna[col].astype('category')

    with mlflow.start_run(run_name='lightgbm_gradient_boosting', tags={'mlflow.user': 'Alexander'}) as run:
        mlflow.log_params({
            "all_days": train_meta["all_days"],
            "train_days_count": train_meta["train_size"],
            "valid_days_count": valid_meta["train_size"],
            "test_days_count": valid_meta["all_days"] - valid_meta["train_size"]
        })

        params = {}
        for cur_model_name, cur_model_col in models.items():
            cur_y_train, cur_y_valid = train[cur_model_col], valid[cur_model_col]

            study = optuna.create_study(study_name=f'Optuna_{cur_model_name}_study', direction='minimize')
            study.optimize(lambda trial: lightgbm_objective(trial, cur_model_name, X_train_optuna, cur_y_train, X_valid_optuna, cur_y_valid), n_trials=50, show_progress_bar=True)

            print(f"--> {cur_model_name} val WAPE = {study.best_value * 100:.2f}%")

            params[cur_model_name] = study.best_params

        del X_train_optuna, X_valid_optuna

        model_pipeline = SklearnPipeline([
            ('preprocessor', DataPreprocessor(exclude_cols, cat_cols)),
            ('clf', MultimodelLGMB(params))
        ])
        with disable_autologging():
            model_pipeline.fit(train, train[targets])
        preds = model_pipeline.predict(test)
        mean_rmse, mean_wape = 0, 0

        for col in targets:
            wape = mean_absolute_percentage_error(test[col], preds[f'pred_{col[7:]}'], sample_weight=test[col])
            rmse = root_mean_squared_error(test[col], preds[f'pred_{col[7:]}'])

            print(f'{col[7:]} predictions\nRMSE = {rmse:.4f}\nWAPE = {wape * 100:.4f}%\n')

            mean_rmse += rmse
            mean_wape += wape

            mlflow.log_metric(f'{col[7:]} RMSE', rmse)
            mlflow.log_metric(f'{col[7:]} WAPE', wape)

        mean_rmse /= len(models)
        mean_wape /= len(models)

        print(f'Mean RMSE = {mean_rmse:.4f}\nMean WAPE = {mean_wape * 100:.4f}%\n')

        mlflow.log_metric('Mean RMSE', mean_rmse)
        mlflow.log_metric('Mean WAPE', mean_wape)

        mlflow.sklearn.log_model(model_pipeline, artifact_path='lightgbm_gb', serialization_format='pickle')
```

```python
lightgbm_train(df)
spark_train(df)

spark.stop()
```

По итогам сравнения `GBTRegressor` оказался самой худшей моделью. Скорее всего это произошло из-за достаточно малого количества данных из-за чего больше времени было потрачено на перестановки датафрейма, а
так же было потрачено намного меньше на тюнинг модели, всего 10 запусков в optuna, вместо 50 у `lightgbm`

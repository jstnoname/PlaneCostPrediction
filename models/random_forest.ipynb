```python
import mlflow
import optuna
from mlflow.utils.autologging_utils import disable_autologging
from optuna import Trial
from pyspark.ml import Pipeline, PipelineModel
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.regression import RandomForestRegressor
from pyspark.sql import SparkSession, DataFrame

from utility_data.utils import time_train_test_split, wape_metric, Config, read_parquet
```

```python
config = Config.load_config()
spark = SparkSession \
    .builder \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.shuffle.partitions", "20") \
    .getOrCreate()

mlflow.set_tracking_uri(config.dagshub_uri)
mlflow.set_experiment('PlaneCostPrediction')
mlflow.pyspark.ml.autolog(log_models=False, log_datasets=False)

df = read_parquet(spark, config.hf_dataset, config.hf_read_token, partition_size=5)

all_features = df.columns
cat_features = [name for name, dtype in df.dtypes if dtype == 'string']

df.show(5)
```

```python
index_cols = [f'{col}_index' for col in cat_features]
exclude_cols = cat_features + ['target_economy', 'target_comfort', 'target_business'] + ['departure_month', 'departure_day']

indexer = StringIndexer(inputCols=cat_features, outputCols=index_cols, handleInvalid='keep')
assembler = VectorAssembler(inputCols=[col for col in all_features if col not in exclude_cols] + index_cols, outputCol='features', handleInvalid='keep')

models = {
    'economy': {'labelCol': 'target_economy', 'featuresCol': 'features', 'predictionCol': 'prediction_economy'},
    'comfort': {'labelCol': 'target_comfort', 'featuresCol': 'features', 'predictionCol': 'prediction_comfort'},
    'business': {'labelCol': 'target_business', 'featuresCol': 'features', 'predictionCol': 'prediction_business'},
}
```

```python
def objective(trial: Trial, model_name: str, model_cols: dict[str, str], train_data: DataFrame, valid_data: DataFrame):
    params = {
        'maxDepth': trial.suggest_int('maxDepth', 2, 10),
        'maxBins': trial.suggest_int('maxBins', 66, 100),
        'numTrees': trial.suggest_int('numTrees', 1, 25),
        'minInfoGain': trial.suggest_float('minInfoGain', 1e-9, 0.1, log=True),
    }

    with disable_autologging():
        forest = RandomForestRegressor(featuresCol=model_cols['featuresCol'], labelCol=model_cols['labelCol'], predictionCol=model_cols['predictionCol'], **params, seed=config.random_state)
        forest = forest.fit(train_data)
    val_preds = forest.transform(valid_data)

    val_wape = wape_metric(val_preds, label_col=model_cols['labelCol'], prediction_col=model_cols['predictionCol'])
    val_rmse = RegressionEvaluator(predictionCol=model_cols['predictionCol'], labelCol=model_cols['labelCol']).evaluate(val_preds)

    mlflow.log_metric(f'{model_name}_val_rmse', val_rmse, step=trial.number)
    mlflow.log_metric(f'{model_name}_val_wape', val_wape, step=trial.number)

    return val_wape
```

```python
with mlflow.start_run(run_name='random_forest', tags={'mlflow.user': 'Alexander'}) as run:
    train, test, train_meta = time_train_test_split(df, 0.25)
    valid, test, valid_meta = time_train_test_split(test, 0.5)

    mlflow.log_params({
        "all_days": train_meta["all_days"],
        "train_days_count": train_meta["train_size"],
        "valid_days_count": valid_meta["train_size"],
        "test_days_count": valid_meta["all_days"] - valid_meta["train_size"]
    })

    print(f"All days = {train_meta['all_days']}\n"
          f"Train days size = {train_meta['train_size']}\n"
          f"Valid days size = {valid_meta['train_size']}\n"
          f"Test days size = {valid_meta['all_days'] - valid_meta['train_size']}")
    print(f"All rows = {df.count()}\n"
          f"Train rows = {train.count()}\n"
          f"Valid rows = {valid.count()}\n"
          f"Test rows = {test.count()}")

    feature_pipeline = Pipeline(stages=[indexer, assembler])
    feature_pipeline = feature_pipeline.fit(train)
    train_piped = feature_pipeline.transform(train).cache()
    valid_piped = feature_pipeline.transform(valid).cache()

    model_stages = []
    for cur_model_name, cur_model_cols in models.items():
        study = optuna.create_study(study_name=f'Optuna_{cur_model_name}_study', direction='minimize')
        study.optimize(lambda trial: objective(trial, cur_model_name, cur_model_cols, train_piped, valid_piped), n_trials=30, show_progress_bar=True)

        print(f"--> {cur_model_name} val WAPE = {study.best_value * 100:.2f}%")

        with disable_autologging():
            model = RandomForestRegressor(
                featuresCol=cur_model_cols['featuresCol'], labelCol=cur_model_cols['labelCol'], predictionCol=cur_model_cols['predictionCol'], **study.best_params, seed=config.random_state
            ).fit(train_piped)
            model_stages.append(model)

    train_piped.unpersist()
    valid_piped.unpersist()

    model_pipeline = PipelineModel(stages=[*feature_pipeline.stages, *model_stages])
    mean_rmse, mean_wape = 0, 0
    preds = model_pipeline.transform(test).cache()

    for cur_model_name, cur_model_cols in models.items():
        rmse = RegressionEvaluator(predictionCol=cur_model_cols['predictionCol'], labelCol=cur_model_cols['labelCol']).evaluate(preds)
        wape = wape_metric(preds, label_col=cur_model_cols['labelCol'], prediction_col=cur_model_cols['predictionCol'])
        print(f'{cur_model_name} predictions\nRMSE = {rmse:.4f}\nWAPE = {wape * 100:.4f}%\n')

        mean_rmse += rmse
        mean_wape += wape

        mlflow.log_metric(f'{cur_model_name} RMSE', rmse)
        mlflow.log_metric(f'{cur_model_name} WAPE', wape)

    mean_rmse /= len(models)
    mean_wape /= len(models)

    print(f'Mean RMSE = {mean_rmse:.4f}\nMean WAPE = {mean_wape * 100:.4f}%\n')

    mlflow.log_metric('Mean RMSE', mean_rmse)
    mlflow.log_metric('Mean WAPE', mean_wape)
    mlflow.spark.log_model(model_pipeline, 'random_forest')
```

Случайный лес показал хороший прирост в метриках, средний WAPE упал с 23% до 15%, средний RMSE упал с 21.4 до 20 билетов. Это произошло из-за того, что случайный лес хорошо ищет нелинейные признаки,
поэтому разброс между классами обслуживания всего 0,359 процента.

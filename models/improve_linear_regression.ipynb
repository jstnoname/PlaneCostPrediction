```python
from pathlib import Path

import mlflow
import optuna
from optuna import Trial
from pyspark.ml import Pipeline, PipelineModel
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml.regression import LinearRegression
from pyspark.sql import SparkSession, DataFrame

from utility_data.utils import time_train_test_split, wape_metric
```

```python
mlflow.set_tracking_uri('http://localhost:5000')
mlflow.set_experiment('PlaneCostPrediction')

spark = SparkSession \
    .builder \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.shuffle.partitions", "20") \
    .getOrCreate()

df_path = Path.cwd().parent / 'data' / 'cleaned_data'
df = spark.read.parquet(str(df_path))

all_features = df.columns
cat_features = [name for name, dtype in df.dtypes if dtype == 'string']

df.show(5)
```

Использовать StringIndexer для замены не численных признаков в линейной регрессии является не плохой практикой, так как модель может решить, что какая-то категория лучше другой, потому что там больше число.
Воспользуемся OHE, для заменя категориальных признаков

```python
index_cols, ohe_cols = map(list, zip(*[(f'{col}_index', f'{col}_ohe') for col in cat_features]))
exclude_cols = index_cols + cat_features + ['target_economy', 'target_comfort', 'target_business']
features = [col for col in all_features if col not in exclude_cols] + ohe_cols

indexer = StringIndexer(inputCols=cat_features, outputCols=index_cols, handleInvalid='keep')
ohe = OneHotEncoder(inputCols=index_cols, outputCols=ohe_cols, handleInvalid='keep')
assembler = VectorAssembler(inputCols=features, outputCol='features', handleInvalid='keep')

models = {
    'economy': {'labelCol': 'target_economy', 'featuresCol': 'features', 'predictionCol': 'prediction_economy'},
    'comfort': {'labelCol': 'target_comfort', 'featuresCol': 'features', 'predictionCol': 'prediction_comfort'},
    'business': {'labelCol': 'target_business', 'featuresCol': 'features', 'predictionCol': 'prediction_business'},
}

mlflow.pyspark.ml.autolog(log_models=False, log_datasets=False)
```

```python
def objective(trial: Trial, model_name: str, model_cols: dict[str, str], train_data: DataFrame, valid_data: DataFrame):
    params = {
        'regParam': trial.suggest_float('regParam', 1e-9, 1, log=True),
        'elasticNetParam': trial.suggest_float('elasticNetParam', 1e-9, 1, log=True),
        'tol': trial.suggest_float('tol', 1e-6, 1e-1, log=True),
        'maxIter': trial.suggest_int('maxIter', 10, 1000),
    }

    with mlflow.start_run(run_name=f'Optuna_{model_name}_trial_{trial.number}', tags={'mlflow.user': 'Alexander'}, nested=True) as run:
        optuna_model = LinearRegression(featuresCol=model_cols['featuresCol'], labelCol=model_cols['labelCol'], predictionCol=model_cols['predictionCol'], **params)
        optuna_model = optuna_model.fit(train_data)
        val_preds = optuna_model.transform(valid_data)

        val_wape = wape_metric(val_preds, label_col=model_cols['labelCol'], prediction_col=model_cols['predictionCol'])
        val_rmse = RegressionEvaluator(predictionCol=model_cols['predictionCol'], labelCol=model_cols['labelCol']).evaluate(val_preds)

        mlflow.log_metric(f'{model_name}_trial_wape', val_wape)
        mlflow.log_metric(f'{model_name}_trial_rmse', val_rmse)

        return val_wape
```

```python
with mlflow.start_run(run_name='improve_linear_regression', tags={'mlflow.user': 'Alexander'}) as run:
    train, test, train_meta = time_train_test_split(df, 0.25)
    valid, test, valid_meta = time_train_test_split(test, 0.5)

    mlflow.log_params({
        "all_days": train_meta["all_days"],
        "train_days_count": train_meta["train_size"],
        "valid_days_count": valid_meta["train_size"],
        "test_days_count": valid_meta["all_days"] - valid_meta["train_size"]
    })

    print(f"All days = {train_meta['all_days']}\n"
          f"Train days size = {train_meta['train_size']}\n"
          f"Valid days size = {valid_meta['train_size']}\n"
          f"Test days size = {valid_meta['all_days'] - valid_meta['train_size']}")
    print(f"All rows = {df.count()}\n"
          f"Train rows = {train.count()}\n"
          f"Valid rows = {valid.count()}\n"
          f"Test rows = {test.count()}")

    features_pipeline = Pipeline(stages=[indexer, ohe, assembler])
    features_pipeline = features_pipeline.fit(train)
    train_piped = features_pipeline.transform(train).cache()
    valid_piped = features_pipeline.transform(valid).cache()

    models_pipeline_stages = []
    for cur_model_name, cur_model_cols in models.items():
        study = optuna.create_study(study_name=f'Optuna_{cur_model_name}_study', direction='minimize')
        study.optimize(lambda trial: objective(trial, cur_model_name, cur_model_cols, train_piped, valid_piped), n_trials=20, show_progress_bar=True)

        print(f"--> {cur_model_name} val WAPE = {study.best_value * 100:.2f}%")

        model = LinearRegression(featuresCol=cur_model_cols['featuresCol'], labelCol=cur_model_cols['labelCol'], predictionCol=cur_model_cols['predictionCol'], **study.best_params).fit(train_piped)
        models_pipeline_stages.append(model)

    train_piped.unpersist()
    valid_piped.unpersist()

    model_pipeline = PipelineModel(stages=[*features_pipeline.stages, *models_pipeline_stages])
    mean_rmse, mean_wape = 0, 0
    preds = model_pipeline.transform(test).cache()

    for model_name, model_cols in models.items():
        rmse = RegressionEvaluator(predictionCol=model_cols['predictionCol'], labelCol=model_cols['labelCol']).evaluate(preds)
        wape = wape_metric(preds, label_col=model_cols['labelCol'], prediction_col=model_cols['predictionCol'])
        print(f'{model_name} predictions\nRMSE = {rmse:.4f}\nWAPE = {wape * 100:.4f}%\n')

        mean_rmse += rmse
        mean_wape += wape

        mlflow.log_metric(f'{model_name} RMSE', rmse)
        mlflow.log_metric(f'{model_name} WAPE', wape)

    mean_rmse /= len(models)
    mean_wape /= len(models)

    print(f'Mean RMSE = {mean_rmse:.4f}\nMean WAPE = {mean_wape * 100:.4f}%\n')

    mlflow.log_metric('Mean RMSE', mean_rmse)
    mlflow.log_metric('Mean WAPE', mean_wape)
```

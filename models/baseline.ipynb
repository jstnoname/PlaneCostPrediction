```python
import mlflow
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.regression import LinearRegression
from pyspark.sql import SparkSession, functions as F

from utility_data.utils import time_train_test_split, wape_metric, Config, read_parquet
```

```python
config = Config.load_config()

spark = SparkSession \
    .builder \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.shuffle.partitions", "20") \
    .getOrCreate()

mlflow.set_tracking_uri(config.dagshub_uri)
mlflow.set_experiment('PlaneCostPrediction')

df = read_parquet(spark, config.hf_dataset, config.hf_read_token, batch_size=10000, partition_size=5)

df.show(5)
```

```python
all_features = df.columns
cat_features = [name for name, dtype in df.dtypes if dtype == 'string']

df \
    .select(cat_features) \
    .agg(*[F.countDistinct(col).alias(col) for col in cat_features]) \
    .show()

df \
    .withColumn('route', F.concat_ws(' -> ', F.concat_ws(' ', 'departure_country', 'departure_city'), F.concat_ws(' ', 'arrival_country', 'arrival_city'))) \
    .select('route') \
    .agg(F.countDistinct(F.col('route')).alias('route_count')) \
    .show()
```

Уникальных значений `airplane_model` немного, всего 8 значений, для линейной модели можно использовать StringIndexer <br>
Остальные же признаки имеют большое количество разных значений. Использовать `route` для baseline'а я не буду, чтобы не утяжелять модель, для `country` можно использовать как OHE, так и StringIndexer,
для `city` я буду использовать StringIndexer, так как их количество достаточно большое <br>
Для baseline'а буду использовать только StringIndexer

Масштаб предсказанных билетов может быть разных, например для маленького регионального рейса 50 билетов будет много, а для большого самолета 50 билетов это очень мало, поэтому необходимо оценивать масштаб
ошибки. Для этого буду использовать метрики `WAPE`, т.к. она показывает в процентах отклонение предсказанного значения от реального, и `RMSE` для понимания насколько сильно модель может отклоняться в
 "пиковых" предсказаниях, который отличаются от настоящих данных

```python
with mlflow.start_run(run_name='baseline_dataset_split_improve', tags={"mlflow.user": "Alexander"}) as run:
    train, test, meta = time_train_test_split(df, 0.25)

    mlflow.log_param('all_days', meta['all_days'])
    mlflow.log_param('train_size', meta['train_size'])

    print(f"All days = {meta['all_days']}\nTrain days size = {meta['train_size']}\nTest days size = {meta['all_days'] - meta['train_size']}")
    print(f"All rows = {df.count()}\nTrain rows = {train.count()}\nTest rows = {test.count()}\n")

    index_cols = [f'{col}_index' for col in cat_features]
    exclude_cols = cat_features + ['target_economy', 'target_comfort', 'target_business']
    feature_cols = [col for col in df.columns if col not in exclude_cols] + index_cols

    indexer = StringIndexer(inputCols=cat_features, outputCols=index_cols, handleInvalid='keep')
    assembler = VectorAssembler(inputCols=feature_cols, outputCol='features', handleInvalid='keep')

    params = {
        "maxIter": 100,
        "regParam": 0.001,
        "elasticNetParam": 0,
        "tol": 0.001,
    }
    models = {
        'economy': {'labelCol': 'target_economy', 'featuresCol': 'features', 'predictionCol': 'prediction_economy'},
        'comfort': {'labelCol': 'target_comfort', 'featuresCol': 'features', 'predictionCol': 'prediction_comfort'},
        'business': {'labelCol': 'target_business', 'featuresCol': 'features', 'predictionCol': 'prediction_business'},
    }

    models_pipeline_stages = []
    for model_name, model_cols in models.items():
        model_params = {**params, **model_cols}
        mlflow.log_params({f"{model_name}_{param_name}": param_value for param_name, param_value in model_params.items()})
        models_pipeline_stages.append(LinearRegression(**model_params))

    pipeline_model = Pipeline(stages=[indexer, assembler, *models_pipeline_stages])
    pipeline_model = pipeline_model.fit(train)
    preds = pipeline_model.transform(test).cache()

    mean_rmse, mean_wape = 0, 0

    for model_name, model_cols in models.items():
        rmse = RegressionEvaluator(predictionCol=model_cols['predictionCol'], labelCol=model_cols['labelCol']).evaluate(preds)
        wape = wape_metric(preds, label_col=model_cols['labelCol'], prediction_col=model_cols['predictionCol'])
        print(f'{model_name} predictions\nRMSE = {rmse:.4f}\nWAPE = {wape * 100:.4f}%\n')

        mean_rmse += rmse
        mean_wape += wape

        mlflow.log_metric(f'{model_name} RMSE', rmse)
        mlflow.log_metric(f'{model_name} WAPE', wape)

    mean_rmse /= len(models)
    mean_wape /= len(models)

    print(f'Mean RMSE = {mean_rmse:.4f}\nMean WAPE = {mean_wape * 100:.4f}%\n')

    mlflow.log_metric('Mean RMSE', mean_rmse)
    mlflow.log_metric('Mean WAPE', mean_wape)
    mlflow.spark.log_model(pipeline_model, 'baseline-linear-regression')
```

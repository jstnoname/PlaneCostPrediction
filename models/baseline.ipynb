```python
from pathlib import Path

import mlflow
from pyspark.sql import SparkSession, functions as F, Window, DataFrame
from pyspark.ml import Pipeline
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
mlflow.set_tracking_uri('http://localhost:5000')
mlflow.set_experiment('PlaneCostPrediction')

spark = SparkSession \
    .builder \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.shuffle.partitions", "20") \
    .getOrCreate()

df_path = Path.cwd().parent / 'data' / 'cleaned_data'
df = spark.read.parquet(str(df_path))

df.show(5)
```

```python
all_features = df.columns
cat_features = [name for name, dtype in df.dtypes if dtype == 'string']

df \
    .select(cat_features) \
    .agg(*[F.countDistinct(col).alias(col) for col in cat_features]) \
    .show()

df \
    .withColumn('route', F.concat_ws(' -> ', F.concat_ws(' ', 'departure_country', 'departure_city'), F.concat_ws(' ', 'arrival_country', 'arrival_city'))) \
    .select('route') \
    .agg(F.countDistinct(F.col('route')).alias('route_count')) \
    .show()
```

Уникальных значений `airplane_model` немного, всего 8 значений, для линейной модели можно использовать StringIndexer <br>
Остальные же признаки имеют большое количество разных значений. Использовать `route` для baseline'а я не буду, чтобы не утяжелять модель, для `country` можно использовать как OHE, так и StringIndexer,
для `city` я буду использовать StringIndexer, так как их количество достаточно большое <br>
Для baseline'а буду использовать только StringIndexer

Масштаб предсказанных билетов может быть разных, например для маленького регионального рейса 50 билетов будет много, а для большого самолета 50 билетов это очень мало, поэтому необходимо оценивать масштаб
ошибки. Для этого буду использовать метрики `WAPE`, т.к. она показывает в процентах отклонение предсказанного значения от реального, и `RMSE` для понимания насколько сильно модель может отклоняться в
 "пиковых" предсказаниях, который отличаются от настоящих данных

```python
def wape_metric(predictions: DataFrame, label_col: str, prediction_col: str):
    errors_sum = predictions \
        .select(
             F.sum(F.abs(F.col(label_col) - F.col(prediction_col))).alias("abs_err_sum"),
            F.sum(F.col(label_col)).alias("actual_sum")
        ) \
        .first()

    if errors_sum is not None and errors_sum.actual_sum != 0:
        return errors_sum.abs_err_sum / errors_sum.actual_sum
    return 0
```

```python
with mlflow.start_run(run_name='Baseline', tags={"mlflow.user": "Alexander"}) as run:
    day_count = df.select('departure_month', 'departure_day').agg(F.count_distinct('departure_month', 'departure_day').alias('day_count')).first().day_count
    train_size = int(.75 * day_count)
    test_size = day_count - train_size

    mlflow.log_param('all_days', day_count)
    mlflow.log_param('train_size', train_size)

    print(f'All days = {day_count}\nTrain days size = {train_size}\nTest days size = {test_size}')

    train = df.withColumn('day', F.dense_rank().over(Window.orderBy('departure_month', 'departure_day'))).filter(F.col('day') <= train_size).drop('day').cache()
    test = df.withColumn('day', F.dense_rank().over(Window.orderBy('departure_month', 'departure_day'))).filter(F.col('day') > train_size).drop('day').cache()

    print(f'All rows = {df.count()}\nTrain rows = {train.count()}\nTest rows = {test.count()}\n')

    index_cols = [f'{col}_index' for col in cat_features]
    exclude_cols = cat_features + ['target_economy', 'target_comfort', 'target_business']
    feature_cols = [col for col in df.columns if col not in exclude_cols] + index_cols

    indexer = StringIndexer(inputCols=cat_features, outputCols=index_cols, handleInvalid='keep')
    assembler = VectorAssembler(inputCols=feature_cols, outputCol='features', handleInvalid='keep')

    params = {
        "maxIter": 100,
        "regParam": 0.001,
        "elasticNetParam": 0,
        "tol": 0.001,
    }
    mlflow.log_params(params)

    model = LinearRegression(featuresCol='features', labelCol='target_economy', predictionCol='prediction', **params)

    pipeline = Pipeline(stages=[indexer, assembler, model])
    pipeline_model = pipeline.fit(train)

    preds = pipeline_model.transform(test)

    rmse = RegressionEvaluator(predictionCol='prediction', labelCol='target_economy').evaluate(preds)
    wape = wape_metric(preds, label_col='target_economy', prediction_col='prediction')

    print(f'RMSE = {rmse:.4f}\nWAPE = {wape * 100:.4f}%')

    mlflow.log_metric('RMSE', rmse)
    mlflow.log_metric('WAPE', wape)
    mlflow.spark.log_model(pipeline_model, 'baseline-linear-regression')
```

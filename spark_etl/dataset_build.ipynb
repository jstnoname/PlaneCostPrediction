````python
from pyspark.sql import SparkSession, functions as F
from pathlib import Path

from utility_data.utils import Config
````

````python
config = Config.load_config()

spark = SparkSession.builder \
    .appName("Flights Analysis") \
    .config("spark.jars.packages", "org.postgresql:postgresql:42.7.3") \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.session.timeZone", "UTC") \
    .config("spark.sql.shuffle.partitions", "20") \
    .getOrCreate()

properties = {
    "user": config.database.username,
    "password": config.database.password,
    "driver": "org.postgresql.Driver",
}
````

Начнем с поиска target'а
немного исследовав базу данных нахожу таблицу segments в которой есть колонки номера билета, его класса обслуживания и цены

````python
df = spark.read.jdbc(url=config.database.url, properties=properties, table='segments', column='flight_id', lowerBound=1, upperBound=35000, numPartitions=10)

df.show(5)
````

````python
target = spark.read.jdbc(url=config.database.url, properties=properties, table='segments', column='flight_id', lowerBound=1, upperBound=35000, numPartitions=10)

target = target \
    .groupBy(target.flight_id) \
    .agg(
        F.count_if(target.fare_conditions == 'Economy').alias("economy"),
        F.count_if(target.fare_conditions == 'Comfort').alias("comfort"),
        F.count_if(target.fare_conditions == 'Business').alias("business"),
    ) \
    .orderBy('flight_id')

target.show(5)
````

из таблицы bookings можно получить book_date, может представлять интерес для понимания за какой срок до вылета бронируются билеты <br>
из tickets атрибут outbound, чтобы понять о том нужен ли билет в обратную сторону <br>
из flights маршруты и время вылета и прилета, чтобы оценить когда люди чаще летают, например днем или ночью <br>
из routes аэропорт отправления и прибытия, чтобы оценить является ли этот маршрут внутреним или международным, большой ли город или нет. Т.к. в routes только коды аэропортов, то соединим с airports_data и получим страну прилета и вылета <br>
из airplanes_data можно скорость самолета, его дальность полета, а если соединить с seats, то и его вместительность по классам обслуживания <br>

Провел анализ данных из базы данных postgres, из данной таблице я получил первую простую таблицу признаков, с которой уже далее можно будет работать <br>
сам sql запрос для извлечения выглядит следующим образом:
```postgresql
select
    flights.flight_id,
    departure.city                                                        as departure_city,
    departure.country                                                     as departure_country,
    array[departure.coordinates[0], departure.coordinates[1]]             as departure_coordinates,
    arrival.city                                                          as arrival_city,
    arrival.country                                                       as arrival_country,
    array[arrival.coordinates[0], arrival.coordinates[1]]                 as arrival_coordinates,
    extract(year from scheduled_departure)                                as departure_year,
    extract(month from scheduled_departure)                               as departure_month,
    extract(day from scheduled_departure)                                 as departure_day,
    extract(isodow from scheduled_departure)                              as departure_day_of_week,
    case
        when extract(isodow from scheduled_departure) in (6,7)
            then true
            else false
    end                                                                   as is_weekend,
    extract(hour from scheduled_departure)                                as departure_hour,
    extract(minute from scheduled_departure)                              as departure_minute,
    extract(hour from duration)                                           as duration_hours,
    extract(minute from duration)                                         as duration_minutes,
    cardinality(days_of_week)                                             as count_of_week_routes,
    airplanes.model                                                       as airplane_model,
    airplanes.range                                                       as airplane_range,
    airplanes.speed                                                       as airplane_speed,
    coalesce(s.economy, 0)                                                as economy,
    coalesce(s.comfort, 0)                                                as comfort,
    coalesce(s.business, 0)                                               as business,
    coalesce(s.all_seats, 0)                                              as all_seats,
    book_data.avg_book_lead_time                                          as avg_book_lead_time,
    book_data.min_book_lead_time                                          as min_book_lead_time,
    book_data.max_book_lead_time                                          as max_book_lead_time,
    book_data.s2_book_lead_time                                           as s2_book_lead_time,
    book_data.count_book_lead_time                                        as count_book_lead_time
from routes
right join flights on
    flights.route_no = routes.route_no and
    flights.scheduled_departure <@ routes.validity
left join (
    select
        flight_id,
        extract(epoch from avg(scheduled_departure - book_date)) / 86400.0    as avg_book_lead_time,
        extract(epoch from min(scheduled_departure - book_date)) / 86400.0    as min_book_lead_time,
        extract(epoch from max(scheduled_departure - book_date)) / 86400.0    as max_book_lead_time,
        stddev(extract(epoch from scheduled_departure - book_date)) / 86400.0 as s2_book_lead_time,
        count(scheduled_departure - book_date)                                as count_book_lead_time
    from flights
    left join segments using(flight_id)
    left join tickets using(ticket_no)
    left join bookings using(book_ref)
    where scheduled_departure > interval '30 days' + book_date
    group by flight_id
) as book_data using(flight_id)
left join airports arrival on arrival.airport_code = routes.arrival_airport
left join airports departure on departure.airport_code = routes.departure_airport
left join airplanes using(airplane_code)
left join (
    select
        airplane_code,
        count(case when seats.fare_conditions = 'Economy' then 1 end)  as economy,
        count(case when seats.fare_conditions = 'Comfort' then 1 end)  as comfort,
        count(case when seats.fare_conditions = 'Business' then 1 end) as business,
        count(*)                                                       as all_seats
    from seats
    group by airplane_code
) as s using(airplane_code)
group by
    flights.flight_id,
    departure_city,
    departure_country,
    departure_coordinates,
    arrival_city,
    arrival_country,
    arrival_coordinates,
    departure_year,
    departure_month,
    departure_day,
    departure_day_of_week,
    is_weekend,
    departure_hour,
    departure_minute,
    duration_hours,
    duration_minutes,
    count_of_week_routes,
    airplane_model,
    airplane_range,
    airplane_speed,
    economy,
    comfort,
    business,
    all_seats,
    avg_book_lead_time,
    min_book_lead_time,
    max_book_lead_time,
    s2_book_lead_time,
    count_book_lead_time
order by flights.flight_id;
```
Теперь реализую данный запрос используя pyspark для сохранения этих данных и будущего использования, так же можно было бы пропустить сохранения и сразу получить эту таблицу, чтобы работать с ней далее.

````python
tickets_subquery = """
(
    select *, abs(hashtext(ticket_no)) % 6000000 as ticket_hash from tickets
) as tickets_subquery
"""
bookings_subquery = """
(
    select *, abs(hashtext(book_ref)) % 2000000 as booking_hash from bookings
) as bookings_subquery
"""

routes = spark.read.jdbc(url=config.database.url, properties=properties, table='routes')
flights = spark.read.jdbc(url=config.database.url, properties=properties, table='flights', column='flight_id', lowerBound=1, upperBound=35000, numPartitions=10)
segments = spark.read.jdbc(url=config.database.url, properties=properties, table='segments', column='flight_id', lowerBound=1, upperBound=35000, numPartitions=10)
tickets = spark.read.jdbc(url=config.database.url, properties=properties, table=tickets_subquery, column='ticket_hash', lowerBound=1, upperBound=5000000, numPartitions=20)
bookings = spark.read.jdbc(url=config.database.url, properties=properties, table=bookings_subquery, column='booking_hash', lowerBound=1, upperBound=2000000, numPartitions=10)
airports = spark.read.jdbc(url=config.database.url, properties=properties, table='airports')
airplanes = spark.read.jdbc(url=config.database.url, properties=properties, table='airplanes')
seats = spark.read.jdbc(url=config.database.url, properties=properties, table='seats')
````

````python
left_bound = F.regexp_extract(routes.validity, r'[\[\(]"?([^,"]+)"?,', 1).cast("timestamp")
right_bound = F.regexp_extract(routes.validity, r',"?([^\]\)]+?)"?[\]\)]', 1).cast("timestamp")
isodow = F.when(F.dayofweek('scheduled_departure') == 1, 7).otherwise(F.dayofweek("scheduled_departure") - 1)
coordinates_parse = F.split(F.regexp_replace(F.col('coordinates'), r'[\(\)]', ''), ',')

aggregated_seats = seats \
    .groupBy(seats.airplane_code) \
    .agg(
        F.count_if(seats.fare_conditions == 'Economy').alias("economy"),
        F.count_if(seats.fare_conditions == 'Comfort').alias("comfort"),
        F.count_if(seats.fare_conditions == 'Business').alias("business"),
        F.count('*').alias('all_seats'),
    )

lead_time = F.unix_timestamp('scheduled_departure') - F.unix_timestamp('book_date')
aggregated_bookings = flights \
    .join(segments, 'flight_id', 'left') \
    .join(tickets, 'ticket_no', 'left') \
    .join(bookings, 'book_ref', 'left') \
    .filter(lead_time > 60 * 60 * 24 * 30) \
    .select('flight_id', (F.unix_timestamp('scheduled_departure') - F.unix_timestamp('book_date')).alias('diff_seconds')) \
    .groupBy('flight_id') \
    .agg(
        (F.avg('diff_seconds') / 86400.0).alias('avg_book_lead_time'),
        (F.min('diff_seconds') / 86400.0).alias('min_book_lead_time'),
        (F.max('diff_seconds') / 86400.0).alias('max_book_lead_time'),
        (F.stddev('diff_seconds') / 86400.0).alias('s2_book_lead_time'),
        F.count('diff_seconds').alias('count_book_lead_time')
    )

departure_airports = airports.select(
    F.col('airport_code').alias('departure_code'),
    F.col('city').alias('departure_city'),
    F.col('country').alias('departure_country'),
    F.array(coordinates_parse[0].cast('double'), coordinates_parse[1].cast('double')).alias('departure_coordinates')
)
arrival_airports = airports.select(
    F.col('airport_code').alias('arrival_code'),
    F.col('city').alias('arrival_city'),
    F.col('country').alias('arrival_country'),
    F.array(coordinates_parse[0].cast('double'), coordinates_parse[1].cast('double')).alias('arrival_coordinates')
)

feature_table = flights \
    .join(
        F.broadcast(routes),
        (flights.route_no == routes.route_no) &
        (flights.scheduled_departure >= left_bound) &
        (flights.scheduled_departure <= right_bound),
        'left'
    ) \
    .join(aggregated_bookings, 'flight_id', 'left') \
    .repartition(20, 'flight_id') \
    .join(F.broadcast(arrival_airports), arrival_airports.arrival_code == routes.arrival_airport, 'left') \
    .join(F.broadcast(departure_airports), departure_airports.departure_code == routes.departure_airport, 'left') \
    .join(F.broadcast(airplanes), 'airplane_code', 'left') \
    .join(F.broadcast(aggregated_seats), 'airplane_code', 'left') \
    .repartition(20, 'flight_id') \
    .select(
        F.col("flight_id"),
        F.col("departure_city"),
        F.col("departure_country"),
        F.col("departure_coordinates"),
        F.col("arrival_city"),
        F.col("arrival_country"),
        F.col("arrival_coordinates"),

        F.year("scheduled_departure").alias("departure_year"),
        F.month("scheduled_departure").alias("departure_month"),
        F.dayofmonth("scheduled_departure").alias("departure_day"),
        isodow.alias("departure_day_of_week"),
        isodow.isin(6, 7).alias("is_weekend"),
        F.hour("scheduled_departure").alias("departure_hour"),
        F.minute("scheduled_departure").alias("departure_minute"),
        F.hour("duration").alias("duration_hours"),
        F.minute("duration").alias("duration_minutes"),
        F.size("days_of_week").alias("count_of_week_routes"),

        F.col("model").alias("airplane_model"),
        F.col("range").alias("airplane_range"),
        F.col("speed").alias("airplane_speed"),

        F.coalesce("economy", F.lit(0)).alias("economy"),
        F.coalesce("comfort", F.lit(0)).alias("comfort"),
        F.coalesce("business", F.lit(0)).alias("business"),
        F.coalesce("all_seats", F.lit(0)).alias("all_seats"),

        F.col("avg_book_lead_time"),
        F.col("min_book_lead_time"),
        F.col("max_book_lead_time"),
        F.col("s2_book_lead_time"),
        F.col("count_book_lead_time")
    ) \
    .orderBy('flight_id')

feature_table.show(5)
````

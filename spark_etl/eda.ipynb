```python
from os import environ
from pathlib import Path
from sys import executable

import numpy as np
import plotly.express as px
import plotly.io as pio
from geopy.distance import geodesic
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.stat import Correlation
from pyspark.sql import SparkSession, functions as F
from pyspark.sql.types import StringType, DoubleType
```

```python
pio.renderers.default = "notebook_connected"

environ['PYSPARK_PYTHON'] = executable
environ['PYSPARK_DRIVER_PYTHON'] = executable

spark = SparkSession \
    .builder \
    .config("spark.driver.memory", "1g") \
    .config("spark.executor.memory", "1g") \
    .config("spark.sql.shuffle.partitions", "20") \
    .getOrCreate()
```

```python
df_path = Path.cwd().parent / "data" / "dataframe"
df = spark.read.parquet(str(df_path))
```

```python
df.show(5)
```

```python
df.printSchema()
```

```python
distance_lambda = F.udf(lambda coord1, coord2: geodesic(coord1[::-1], coord2[::-1]).km, DoubleType())

df = df \
    .withColumn('distance', distance_lambda('departure_coordinates', 'arrival_coordinates')) \
    .drop('departure_coordinates', 'arrival_coordinates')
```

```python
df.printSchema()
```

```python
all_features = df.columns
cat_features = [feature.name for feature in df.schema if isinstance(feature.dataType, StringType)]
noncat_features = [feature.name for feature in df.schema if not isinstance(feature.dataType, StringType)]

print(f'Categorical features: {cat_features}')
print(f'Non-categorical features: {noncat_features}')

bool_replace = [col for col, dtype in df.dtypes if dtype == 'boolean']
df = df.withColumns({col: F.col(col).cast('int') for col in bool_replace})
```

```python
null_and_nan_cols = [
    col
    for col, count in df \
        .select(
            [F.count(F.when(F.isnull(col) | F.isnan(col), col)).alias(col) for col in noncat_features] +
            [F.count(F.when(F.isnull(col), col)).alias(col) for col in cat_features]
        ) \
        .first() \
        .asDict() \
        .items()
    if count > 0
]

print(f'cols with null or nan values: {null_and_nan_cols}')
```

т.к. физический смысл данных может иметь разное значение, то и обрабатывать каждую из этих колонок надо отдельно <br>
`count_book_lead_time` заменить на 0, потому что null или nan означает отсутствие броней, то есть 0 <br>
для остальный колоном можно добавить колонку `заменено ли это значение` и заменить на 0, чтобы в будущей модель лучше понимала реальное ли это значение или нет <br>
замена на 0 происходит из-за того, что это данные уже динамические, которые считаются из количества билетов, которые люди купили за 30 дней до рейса. Заменять их на другие значение означает, что мы даем не реальные данные

```python
for col in null_and_nan_cols:
    if col == 'count_book_lead_time':
        df = df.fillna({col: 0})
    else:
        df = df.withColumn(f"is_{col}_missing", F.when(F.isnan(col) | F.isnull(col), 1).otherwise(0))   # все колонки это числовые значение
        df = df.fillna({col: 0})

df.show(5)
```

```python
df.printSchema()
```

работа с пропусками и со значениями типа boolean завершена <br>
теперь можно переходить к самому анализу данных

```python
target_classes = df \
    .select(
        F.sum("target_economy").alias("economy"),
        F.sum("target_comfort").alias("comfort"),
        F.sum("target_business").alias("business")
    )
fig = px.bar(x=target_classes.columns, y=target_classes.collect(), title='Fare Conditions Distribution', labels={'x': 'Ticket classes', 'y': 'Ticket amount'}).update_layout(showlegend=False)
fig.show()
```

![Fare distribution](https://github.com/jstnoname/PlaneCostPrediction/blob/main/graphs/fare_distributions.png?raw=true)

по графику видим, что за все время эконом класс купили почти 6 миллионов раз, в то время, как комфорт и бизнес меньше 1 миллиона

```python
describe_agg = lambda col: (
    F.count(col).alias(f"{col}_count"),
    F.min(col).alias(f"{col}_min"),
    F.max(col).alias(f"{col}_max"),
    F.avg(col).alias(f"{col}_avg"),
    F.mean(col).alias(f"{col}_mean"),
    F.stddev(col).alias(f"{col}_stddev"),
)

df \
    .withColumn('route', F.concat_ws(' -> ', F.concat_ws(' ', 'departure_country', 'departure_city'), F.concat_ws(' ', 'arrival_country', 'arrival_city'))) \
    .withColumn('total_tickets', F.col('target_economy') + F.col('target_comfort') + F.col('target_business')) \
    .withColumn('load_factor', F.col('total_tickets') / F.col('all_seats')) \
    .groupBy('route') \
    .agg(
        *describe_agg('target_economy'),
        *describe_agg('target_comfort'),
        *describe_agg('target_business'),
        F.avg('load_factor').alias('avg_load_factor'),
        F.max('load_factor').alias('max_load_factor'),
        F.min('load_factor').alias('min_load_factor')
    ) \
    .show()
```

```python
overbooking_df = df.filter(
    (F.col("target_economy") > F.col("economy")) |
    (F.col("target_comfort") > F.col("comfort")) |
    (F.col("target_business") > F.col("business"))
).count()

print(f"Overbooking flights: {overbooking_df}")
```

По этим данным можно сказать, что количество мест зависит от типа самолета и его вместительности и маршрута <br>
в данной таблице добавил колонку загрузки самолета для оценки относительно маршрута <br>
теперь надо сделать то же самое но для времени года для проверки сезонности. Возьму для группировки месяц вылета, так как группировать для каждого дня месяца может быть очень тяжело, так же для данной группы может быть мало данных, а в группировке по году будет слишком большое количество данных для выделения какой-нибудь закономерности

```python
df \
    .withColumn('total_tickets', F.col('target_economy') + F.col('target_comfort') + F.col('target_business')) \
    .withColumn('load_factor', F.col('total_tickets') / F.col('all_seats')) \
    .groupBy('departure_year', 'departure_month') \
    .agg(
        F.count('flight_id').alias('flights'),
        F.avg('load_factor').alias('avg_load_factor'),
        F.max('load_factor').alias('max_load_factor'),
        F.min('load_factor').alias('min_load_factor')
    ) \
    .orderBy('departure_year', 'departure_month') \
    .show()
```

По таблице видно, что это данные непрерывны в течении 7 месяцев, начиная с октября и заканчивая апрелем. Явным аутсайдером является апрель, средняя загрузка самолета меньше половины и количество полетов в этот месяц так же меньше, чем в другие. Так же в марте загрузка самолетов меньше, чем раньше, но не столь критично как в апреле. <br>
Значит существует зависимость полетов и времени года

```python
indexer = StringIndexer(inputCol='airplane_model', outputCol='airplane_model_index')
df = indexer.fit(df).transform(df)
noncat_features = [feature.name for feature in df.schema if not isinstance(feature.dataType, StringType) and feature.name != 'flight_id']

assembler = VectorAssembler(inputCols=noncat_features, outputCol='corr_features')
df_vector = assembler.transform(df)
matrix = Correlation.corr(df_vector, 'corr_features').head()
dense_matrix = np.round(matrix.asDict()['pearson(corr_features)'].toArray(), 3)
```

```python
fig = px.imshow(
    dense_matrix,
    x=noncat_features,
    y=noncat_features,
    color_continuous_scale='RdBu',
    range_color=(-1, 1),
    aspect='auto',
    title='Correlation Matrix',
)
fig.show()
```

![Матрица корреляции](https://github.com/jstnoname/PlaneCostPrediction/blob/main/graphs/corr_matrix.png?raw=true)


по данной матрице видно, что наши таргеты коррелируют с дистанцией, общим количество мест (лимит сверху) и количество уже забронированных билетов <br>
так же из предыдущих расчетов мы выяснили, что время года, маршрут, модель самолета и соответственно дистанция коррелирует с количеством проданных билетов <br>
уберем признаки, которые сильно коррелируют с другими, чтобы в будущем модель лучше искала зависимости

```python
target = [
    'target_economy',
    'target_comfort',
    'target_business',
]

# оставим только те признаки, которые нам будут нужны для будущих моделей. Как линейных, так и нет
features = [
    # временные
    'departure_day',
    'departure_day_of_week',
    'departure_month',
    'is_weekend',
    'departure_hour',

    # физический смысл рейса
    'duration_hours',
    'count_of_week_routes',
    'airplane_range',
    'economy',
    'comfort',
    'business',

    # метрики бронирования за 30 дней
    'avg_book_lead_time',
    's2_book_lead_time',
    'count_book_lead_time',

    # метрики пропуска данных
    'is_avg_book_lead_time_missing'
]

cat_features = [
    'departure_city',
    'departure_country',
    'arrival_city',
    'arrival_country',
    'airplane_model'
]
```

```python
final_columns = target + features + cat_features
final_columns
```

```python
df = df.select(final_columns)
df.show(5)
```
